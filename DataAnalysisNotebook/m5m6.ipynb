###Some dependencies to inherit for running the tests.

```
from scipy.stats import spearmanr
import matplotlib.pyplot as plt
```

```
#Software defect density results Collections
software_defect_density_results = [ 0.5106, 0.1006, 0.1272, 0.2005, 0.0449,0.05966]
#Adaptive Maintenance Effort Results
adaptive_maintenance_effort_results = [ 88.15, 91.09, 121.33, 191.2, 133.97, 89.37]
```

```
print(spearmanr(software_defect_density_results, 
                adaptive_maintenance_effort_results))
```

###Due to the small sample size, the significance level is about 46% and hence we can't reject the null hypothesis. 

```
software_defect_density_results,adaptive_maintenance_effort_results = zip(*sorted(zip( software_defect_density_results, 
                                                                                   adaptive_maintenance_effort_results )))
plt.plot(software_defect_density_results, adaptive_maintenance_effort_results)
plt.xlabel("Software defect density")
plt.ylabel("Adaptive maintenance effort")
plt.show()
```
